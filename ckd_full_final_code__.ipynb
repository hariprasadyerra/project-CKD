### Data Cleaning


```python
# Data Cleaning

# Standardize column names → lowercase, snake_case
df.columns = (
    df.columns
      .str.strip()                                    # remove extra spaces
      .str.lower()                                   # lowercase
      .str.replace(r'[^0-9a-z]+', '_', regex=True)   # replace non-alphanumeric with _
      .str.replace(r'_+', '_', regex=True)           # collapse multiple underscores
      .str.strip('_')                                # remove leading/trailing _
)

#  Replace common missing tokens with NaN
MISSING_TOKENS = ['?', 'na', 'n/a', 'nil', 'null', ' ', '\t?']
df = df.replace(MISSING_TOKENS, pd.NA)

#  Clean text columns  spaces, remove tabs
obj_cols = df.select_dtypes(include='object').columns
for c in obj_cols:
    df[c] = df[c].str.strip().str.replace('\t', '', regex=False)

#  Drop irrelevant id column
if 'id' in df.columns:
    df = df.drop(columns=['id'])

#   check after cleaning
print("Shape of dataset:", df.shape)
print("\nColumn names:", list(df.columns))

print("Missing values (count)")
print(df.isna().sum())

print("Head (post-clean)")
display(df.head())
```

```python
#Target Preparation

#  Clean classification column text
df['classification'] = df['classification'].astype(str).str.strip().str.lower()
df['classification'] = df['classification'].replace({'ckd\t': 'ckd'})

#  Map to numeric: 1 = CKD, 0 = Not CKD
df['target_ckd'] = df['classification'].map({'ckd': 1, 'notckd': 0}).astype('Int64')

#  Drop the classification text column
df = df.drop(columns=['classification'])

#  Check target distribution
print("Shape after target prep:", df.shape)
print("nTarget distribution (0 = Not CKD, 1 = CKD)")
print(df['target_ckd'].value_counts(dropna=False))

print("Head after target prep:")
display(df.head())
```

### encoding text to numeric

```python
#  Categorical Encoding
#  convert medical text fields to numeric (0/1) and ensure pcv/wc/rc are numeric
import pandas as pd
import numpy as np

#  Define mappings
yes_no_map = {'yes': 1, 'no': 0}
normal_abnormal_map = {'normal': 1, 'abnormal': 0}
present_map = {'present': 1, 'notpresent': 0}
appet_map = {'good': 1, 'poor': 0}

#  Apply mappings where columns exist
mapping_plan = {
    'htn': yes_no_map,
    'dm': yes_no_map,
    'cad': yes_no_map,
    'pe': yes_no_map,
    'ane': yes_no_map,
    'rbc': normal_abnormal_map,
    'pc': normal_abnormal_map,
    'pcc': present_map,
    'ba': present_map,
    'appet': appet_map
}

for col, mp in mapping_plan.items():
    if col in df.columns:
        # standardize any lingering strings (safety)
        df[col] = df[col].astype(str).str.strip().str.lower()
        df[col] = df[col].replace({'nan': np.nan})  # keep true NaN as NaN
        df[col] = df[col].map(mp)

# Ensure medical numeric strings are numeric
for col in ['pcv', 'wc', 'rc']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

#  cheking values
print("Unique values after encoding (should be 2 for binaries):")
bin_cols = [c for c in ['htn','dm','cad','pe','ane','rbc','pc','pcc','ba','appet'] if c in df.columns]
if bin_cols:
    print(df[bin_cols].nunique())

print("\nDtypes summary:")
print(df.dtypes)

print("Head ")
display(df.head())
```

###Outlier Handling using IQR

```python

#  Outlier Handling using IQR (Capping) + Grouped Plots

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

#pick continuous numeric features (exclude target & binary-like)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if "target_ckd" in numeric_cols:
    numeric_cols.remove("target_ckd")
continuous_cols = [c for c in numeric_cols if df[c].nunique(dropna=True) > 2]

print("Continuous features for outlier treatment:", continuous_cols)

#  IQR capping
def cap_outliers_iqr(data, cols, k=1.5):
    capped = data.copy()
    for col in cols:
        Q1 = capped[col].quantile(0.25)
        Q3 = capped[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - k * IQR
        upper = Q3 + k * IQR
        capped[col] = capped[col].clip(lower=lower, upper=upper)
    return capped

# grouped boxplots BEFORE (3 per row)
def grouped_boxplots(data, cols, title, color):
    n = len(cols)
    cols_per_row = 3
    rows = (n + cols_per_row - 1) // cols_per_row
    fig, axes = plt.subplots(rows, cols_per_row, figsize=(15, 4*rows))
    axes = np.array(axes).reshape(rows, cols_per_row)
    for i, col in enumerate(cols):
        r, c = divmod(i, cols_per_row)
        sns.boxplot(x=data[col], ax=axes[r, c], color=color)
        axes[r, c].set_title(col)
        axes[r, c].set_xlabel(col)
        axes[r, c].set_ylabel("Distribution")
    # hide empty slots
    for j in range(i+1, rows*cols_per_row):
        r, c = divmod(j, cols_per_row)
        axes[r, c].axis("off")
    plt.suptitle(title, fontsize=14, y=1.02)
    plt.tight_layout()
    plt.show()

print("Boxplots — BEFORE capping")
grouped_boxplots(df, continuous_cols, "Before Outlier Capping (IQR)", color="skyblue")

# apply capping (k=1.5)
rows_before = df.shape[0]
class_before = df['target_ckd'].value_counts()

df_capped = cap_outliers_iqr(df, continuous_cols, k=1.5)

rows_after = df_capped.shape[0]
class_after = df_capped['target_ckd'].value_counts()

print("Row count (unchanged)", rows_before, "→", rows_after)
print("Class distribution (unchanged):")
print("Before:\n", class_before)
print("After:\n", class_after)

print("Boxplots — AFTER capping")
grouped_boxplots(df_capped, continuous_cols, "After Outlier Capping (IQR)", color="lightgreen")

# use capped data
df = df_capped.copy()
print("Outlier capping complete. Using capped dataset for next steps.")
```

```python

#  Feature Engineering + Train/Test Split + SMOTE
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import seaborn as sns
import matplotlib.pyplot as plt


# Split into Features (X) and Target (y)

# Drop the temporary plotting column '_target_label' if it exists from features
cols_to_drop_X = ['target_ckd']
if '_target_label' in df.columns:
    cols_to_drop_X.append('_target_label')

X = df.drop(columns=cols_to_drop_X)
y = df['target_ckd']

# Split into train and test (keep target distribution same)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=21
)

print("Shapes after split:")
print("X_train:", X_train.shape, "| y_train:", y_train.shape)
print("X_test :", X_test.shape,  "| y_test :", y_test.shape)


#  Handle Missing Values

# Continuous (numeric) columns
numeric_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
continuous_cols = [c for c in numeric_cols if X_train[c].nunique() > 2]

# Imputer (fit only on train, apply to test)
imputer = SimpleImputer(strategy="median")
X_train[continuous_cols] = imputer.fit_transform(X_train[continuous_cols])
X_test[continuous_cols]  = imputer.transform(X_test[continuous_cols])


#  Fix Skewness (only on train to decide)

skew_vals = X_train[continuous_cols].skew().sort_values(ascending=False)
highly_skewed = skew_vals[abs(skew_vals) > 1].index.tolist()

print("\nSkewness (train):")
print(skew_vals)
print("Applying log1p to:", highly_skewed)

for col in highly_skewed:
    # log1p needs non-negative values
    X_train[col] = np.where(X_train[col] < 0, 0, X_train[col])
    X_test[col]  = np.where(X_test[col] < 0, 0, X_test[col])
    X_train[col] = np.log1p(X_train[col])
    X_test[col]  = np.log1p(X_test[col])


#  Scale Features

scaler = StandardScaler()
X_train[continuous_cols] = scaler.fit_transform(X_train[continuous_cols])
X_test[continuous_cols]  = scaler.transform(X_test[continuous_cols])

print(" Missing values handled, skewed features log-transformed, and scaling done.")


# Handle Class Imbalance (SMOTE)

smote = SMOTE(random_state=21)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

print("\nShapes after SMOTE:")
print("X_train:", X_train_sm.shape, "| y_train:", y_train_sm.shape)
print("X_test :", X_test.shape,     "| y_test :", y_test.shape)

print("\nBalanced class distribution (training set only):")
print(y_train_sm.value_counts())


# Show Class Balance Before and After SMOTE

plt.figure(figsize=(6,4))
sns.countplot(x=y_train, palette="Set2")
plt.title("Before SMOTE (Train)")
plt.xlabel("CKD (0 = No, 1 = Yes)")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x=y_train_sm, palette="Set2")
plt.title("After SMOTE (Train)")
plt.xlabel("CKD (0 = No, 1 = Yes)")
plt.show()
```

```python
#  KNN implementing

# we train on the SMOTE-balanced training set (X_train_sm, y_train_sm)
# we always test on the original test set (X_test, y_test)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    confusion_matrix, classification_report, recall_score,
    roc_auc_score, average_precision_score, roc_curve, precision_recall_curve
)


#  Train the KNN model
knn_base = KNeighborsClassifier(
    n_neighbors=5,      # simple default
    weights="distance", # nearer neighbors matter more
    p=2                 # Euclidean distance
)
knn_base.fit(X_train_sm, y_train_sm)

# Predict on the TEST set
y_pred = knn_base.predict(X_test)
y_proba = knn_base.predict_proba(X_test)[:, 1]


# Print metrics
print(" KNN (model) — TEST RESULTS ")
print("Recall:", round(recall_score(y_test, y_pred), 3))         # most important in health
print("ROC AUC:", round(roc_auc_score(y_test, y_proba), 3))
print("PR  AUC:", round(average_precision_score(y_test, y_proba), 3))
print("\nClassification report:\n", classification_report(y_test, y_pred, digits=3))
```

####  Hyperparameter Tuning

```python

#  Hyperparameter Tuning (All 4 Models)
# (KNN, LightGBM, Logistic Regression, Naive Bayes)
import numpy as np
import pandas as pd

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, average_precision_score, classification_report
)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from lightgbm import LGBMClassifier


#  Ensure consistent feature names (no warnings)
feature_names = X_train.columns.tolist()
X_train_sm_df = pd.DataFrame(X_train_sm, columns=feature_names)
X_test_df     = pd.DataFrame(X_test,     columns=feature_names)


#  KNN — GridSearchCV (scoring = recall)
knn = KNeighborsClassifier()
knn_grid = {
    "n_neighbors": [3, 5, 7, 9, 11],
    "weights": ["uniform", "distance"],
    "p": [1, 2]   # 1=Manhattan, 2=Euclidean
}
gs_knn = GridSearchCV(knn, knn_grid, cv=5, scoring="recall", n_jobs=-1, verbose=0)
gs_knn.fit(X_train_sm_df, y_train_sm)
best_knn = gs_knn.best_estimator_

# Predictions
y_pred_knn  = best_knn.predict(X_test_df)
y_proba_knn = best_knn.predict_proba(X_test_df)[:, 1]

# LightGBM — GridSearchCV
lgbm = LGBMClassifier(random_state=21, verbosity=-1)
lgbm_grid = {
    "n_estimators": [100, 200, 300],
    "learning_rate": [0.01, 0.05, 0.1],
    "num_leaves": [15, 31, 63],
    "max_depth": [-1, 5, 10]
}
gs_lgbm = GridSearchCV(lgbm, lgbm_grid, cv=5, scoring="recall", n_jobs=-1, verbose=0)
gs_lgbm.fit(X_train_sm_df, y_train_sm)
best_lgbm = gs_lgbm.best_estimator_

y_pred_lgbm  = best_lgbm.predict(X_test_df)
y_proba_lgbm = best_lgbm.predict_proba(X_test_df)[:, 1]

#  Logistic Regression — GridSearchCV
log = LogisticRegression(max_iter=1000, random_state=21)
log_grid = {
    "C": [0.01, 0.1, 1, 10],
    "penalty": ["l2"],          # use l2 with lbfgs
    "solver": ["lbfgs"]
}
gs_log = GridSearchCV(log, log_grid, cv=5, scoring="recall", n_jobs=-1, verbose=0)
gs_log.fit(X_train_sm_df, y_train_sm)
best_log = gs_log.best_estimator_

y_pred_log  = best_log.predict(X_test_df)
y_proba_log = best_log.predict_proba(X_test_df)[:, 1]


#  Naive Bayes (GaussianNB) — GridSearchCV
nb = GaussianNB()
nb_grid = {
    "var_smoothing": [1e-9, 1e-8, 1e-7, 1e-6]
}
gs_nb = GridSearchCV(nb, nb_grid, cv=5, scoring="recall", n_jobs=-1, verbose=0)
gs_nb.fit(X_train_sm_df, y_train_sm)
best_nb = gs_nb.best_estimator_

y_pred_nb  = best_nb.predict(X_test_df)
y_proba_nb = best_nb.predict_proba(X_test_df)[:, 1]

# Summary table of tuned models (TEST set)
def summarize(name, y_true, y_pred, y_proba):
    return {
        "Model": name,
        "Accuracy":  round(accuracy_score(y_true, y_pred), 3),
        "Precision": round(precision_score(y_true, y_pred, zero_division=0), 3),
        "Recall":    round(recall_score(y_true, y_pred), 3),
        "F1":        round(f1_score(y_true, y_pred), 3),
        "ROC AUC":   round(roc_auc_score(y_true, y_proba), 3),
        "PR  AUC":   round(average_precision_score(y_true, y_proba), 3)
    }

tuned_results = [
    summarize("KNN (tuned)",             y_test, y_pred_knn,  y_proba_knn),
    summarize("LightGBM (tuned)",        y_test, y_pred_lgbm, y_proba_lgbm),
    summarize("Logistic (tuned)",        y_test, y_pred_log,  y_proba_log),
    summarize("Naive Bayes (tuned)",     y_test, y_pred_nb,   y_proba_nb),
]
tuned_df = pd.DataFrame(tuned_results)
print(" Tuned Models — TEST Metrics ")
print(tuned_df)


# Display results for each tuned model
def show_results(name, y_true, y_pred, y_proba, grid):
    print(f" {name}")
    print("Best Parameters:", grid.best_params_)
    print("Accuracy :", round(accuracy_score(y_true, y_pred), 3))
    print("Precision:", round(precision_score(y_true, y_pred, zero_division=0), 3))
    print("Recall   :", round(recall_score(y_true, y_pred), 3))
    print("F1 Score :", round(f1_score(y_true, y_pred), 3))
    print("ROC AUC  :", round(roc_auc_score(y_true, y_proba), 3))
    print("PR  AUC  :", round(average_precision_score(y_true, y_proba), 3))
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, digits=3, zero_division=0))

#  Run for each model
show_results("KNN (Tuned)", y_test, y_pred_knn, y_proba_knn, gs_knn)
show_results("LightGBM (Tuned)", y_test, y_pred_lgbm, y_proba_lgbm, gs_lgbm)
show_results("Logistic Regression (Tuned)", y_test, y_pred_log, y_proba_log, gs_log)
show_results("Naive Bayes (Tuned)", y_test, y_pred_nb, y_proba_nb, gs_nb)

```

```python
# Plots for Tuned Models
# Confusion Matrix + ROC + Precision–Recall
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    confusion_matrix, roc_curve, auc,
    precision_recall_curve, average_precision_score
)

# Helper: one model → 3 plots (CM, ROC, PR)
def plot_model_performance(y_true, y_pred, y_proba, model_name):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    #  Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, ax=axes[0])
    axes[0].set_title(f"{model_name} — Confusion Matrix")
    axes[0].set_xlabel("Predicted")
    axes[0].set_ylabel("Actual")

    #  ROC Curve
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    roc_auc = auc(fpr, tpr)
    axes[1].plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
    axes[1].plot([0, 1], [0, 1], linestyle="--")
    axes[1].set_title(f"{model_name} — ROC Curve")
    axes[1].set_xlabel("False Positive Rate")
    axes[1].set_ylabel("True Positive Rate")
    axes[1].legend(loc="lower right")
    axes[1].grid(True)

    #  Precision–Recall Curve
    prec, rec, _ = precision_recall_curve(y_true, y_proba)
    pr_auc = average_precision_score(y_true, y_proba)
    axes[2].plot(rec, prec, label=f"PR AUC = {pr_auc:.2f}")
    axes[2].set_title(f"{model_name} — Precision–Recall Curve")
    axes[2].set_xlabel("Recall")
    axes[2].set_ylabel("Precision")
    axes[2].legend(loc="lower left")
    axes[2].grid(True)

    plt.tight_layout()
    plt.show()



# Build predictions/probabilities (TEST)
# KNN
y_pred_knn  = best_knn.predict(X_test_df)
y_proba_knn = best_knn.predict_proba(X_test_df)[:, 1]

# LightGBM
y_pred_lgbm  = best_lgbm.predict(X_test_df)
y_proba_lgbm = best_lgbm.predict_proba(X_test_df)[:, 1]

# Logistic Regression
y_pred_log  = best_log.predict(X_test_df)
y_proba_log = best_log.predict_proba(X_test_df)[:, 1]

# Naive Bayes
y_pred_nb  = best_nb.predict(X_test_df)
y_proba_nb = best_nb.predict_proba(X_test_df)[:, 1]



# Plot each tuned model
plot_model_performance(y_test, y_pred_knn,  y_proba_knn,  "KNN (Tuned)")
plot_model_performance(y_test, y_pred_lgbm, y_proba_lgbm, "LightGBM (Tuned)")
plot_model_performance(y_test, y_pred_log,  y_proba_log,  "Logistic Regression (Tuned)")
plot_model_performance(y_test, y_pred_nb,   y_proba_nb,   "Naive Bayes (Tuned)")
```
